```python
#Draw the line chart using “x” as the output given here
#import numpy as np
#x = np.arange(11)

import numpy as np
import matplotlib.pyplot as plt

# Data
x = np.arange(11)  # 0 to 10
y1 = x + 2
y2 = x + 3

# Plotting
plt.plot(x, y1, label="x+2")
plt.plot(x, y2, label="x+3")

# Labels and Legend
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title("Line Chart of x+2 and x+3")
plt.show()
```
